```python
#b The difference between Deep copy and shallow copy
```

1. Shallow Copy
A shallow copy creates a new object but does not recursively copy the objects contained within the original object. Instead, it copies references to the original objects. So, if the original object contains other mutable objects, the shallow copy will reference those same objects.

Only the top-level object is copied.
Changes to mutable objects inside the original will affect the shallow copy (since both reference the same inner objects).

```python
import copy

original_list = [[1, 2, 3], [4, 5, 6]]
shallow_copy = copy.copy(original_list)
original_list[0][0] = 99
print("Original List:", original_list)
print("Shallow Copy:", shallow_copy)
```

2. Deep Copy
A deep copy creates a new object and recursively copies all objects it contains. This means that not only is the top-level object copied, but all nested objects are copied as well, ensuring that the original and copied objects are completely independent.

Both the top-level object and inner objects are copied.
Changes to any objects inside the original do not affect the deep copy.

```python
import copy

original_list = [[1, 2, 3], [4, 5, 6]]
deep_copy = copy.deepcopy(original_list)

original_list[0][0] = 99
print("Original List:", original_list)
print("Deep Copy:", deep_copy)
```

```python
#14 Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' .
import pandas as pd
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')
temperature = np.random.normal(loc=20, scale=5, size=100)
humidity = np.random.normal(loc=60, scale=10, size=100)

df = pd.DataFrame({'Date': date_range, 'Temperature': temperature, 'Humidity': humidity})

#a)Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

#b)  Label the x-axis as 'Date
ax1.set_xlabel('Date')

#c)  Set the title of the plot as 'Temperature and Humidity Over Time'.
plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python
#15 Create a NumPy array data containing 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

#a) Plot a histogram of the data with 30 bins.
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

#b) Overlay a line plot representing the normal distribution's probability density function (PDF).
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

mean, std = data.mean(), data.std()
pdf = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std) ** 2)

plt.plot(x, pdf, color='red', linewidth=2, label='Normal Distribution PDF')

#c)  Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

#) Set the title of the plot as 'Histogram with PDF Overlay'.
plt.title('Histogram with PDF Overlay')
```

```python
#17  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
#origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

x_rand = np.random.uniform(-1, 1, 100)
y_rand = np.random.uniform(-1, 1, 100)

colors = ['red' if x > 0 and y > 0 else 'green' if x < 0 and y > 0 else 'blue' if x < 0 and y < 0 else 'purple' for x, y in zip(x_rand, y_rand)]

plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_rand, y=y_rand, hue=colors, palette={'red': 'red', 'green': 'green', 'blue': 'blue', 'purple': 'purple'}, s=50)

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(['Quadrant I', 'Quadrant II', 'Quadrant III', 'Quadrant IV'], title='Quadrant', loc='upper right')
plt.show()
```

```python
#18 8 With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'8
from bokeh.plotting import figure, show
import numpy as np

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

p.line(x, y, line_width=2, color="blue")

p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"

show(p)
```

```python
#19 Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart?
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(1, 100, size=len(categories))

source = ColumnDataSource(data=dict(categories=categories, values=values, color=values))

p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', tools="")

bars = p.vbar(x='categories', top='values', width=0.9, source=source,
              color='color', legend_field='categories')

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

p.xgrid.grid_line_color = None
p.legend.orientation = "vertical"
p.legend.location = "top_left"
show(p)
```

```python
#20 Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'
import plotly.graph_objs as go
import numpy as np

x = np.linspace(0, 10, 100)
y = np.random.randn(100)

fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

fig.update_layout(title='Simple Line Plot',
                  xaxis_title='X-axis',
                  yaxis_title='Y-axis')

fig.show()
```
